```python
import pandas as pd
import numpy as np

# Creating a DataFrame from dictionary
data = {
    'student_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'hours_studied': [65, 3, 8, 2, 6, 65],
    'previous_scores': [85, 70, 90, 50, 60, 50],
    'attendance': [50, 30, 85, 45, 22, 50],
    'sleepy_hours': [7.5, 3.8, 8.2, 4.5, 6.5, 2.3],
    'extracurricular': ['yes', 'no', 'yes', 'no', 'yes', 'no'],
    'parent_education': ['bachelor', 'high school', 'master', 'PhD', 'bachelor', 'bachelor']
}

df = pd.DataFrame(data)

print("Student Performance DataFrame")
print(df)

print("\nData Info:")
df.info()

# Basic DataFrame exploration
print("\nDataFrame shape:", df.shape[0], "rows,", df.shape[1], "columns")
print("DataFrame columns:", df.columns.tolist())
print("Datatypes:\n", df.dtypes)
```

```python
# View first 5 rows
print("First 5 rows:")
print(df.head())

# View last 2 rows
print("\nLast 2 rows:")
print(df.tail(2))

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check for NaN values
print("\nNaN Values:")
print(df.isna().sum())

# Check for duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())

# Remove duplicate rows
print("\nRemoving Duplicates:")
df = df.drop_duplicates()

# Display DataFrame after removing duplicates
df.drop_duplicates(inplace=True)
print(df)
```

```python
def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
    
    df['grade'] = df['previous_scores'].apply(calculate_grade)
print("\nDataFrame with new columns:,", df)

#using map for categorical data.
extracurricular_mapp = {'yes': 1, 'no': 0}
df['extracurricular_binary'] = df['extracurricular'].map(extracurricular_mapp)
```
